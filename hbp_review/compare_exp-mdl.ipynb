## Initialization

```python
import sys
sys.path.insert(1, '..')

# change this to point to your reachgrasp IO
sys.path.insert(1, '../../dataset_repos/reachgrasp/python')
sys.path.insert(1, '../../toolboxes/py/python-neo')
sys.path.insert(1, '../../toolboxes/py/python-odml')

# standard python library
import pickle

import numpy as np
import scipy
from quantities import s, ms, Hz
import matplotlib.pyplot as plt
%matplotlib inline

# data model for ephys data
import neo

# core analysis library component
import elephant

# provides plots of the results
import plots
```

Global variables: offset and duration of recording to load

```python
rec_start = 10. * s
duration = 50. * s
bin_size=20. * ms
```

Colors for experiment and model

```python
exp_col=[1., 0.6, 0.4]
mdl_col=[0.6, 0.4, 1.]
```

## Load experimental data

Load experimental data using the neo HDF5 IO.

```python
filename = 'data/experiment.h5'
session = neo.NeoHdf5IO(filename=filename)
block_exp = session.read_block()
```

Select spike trains (min. 2 spikes, SUA only).

```python
sts_exp = [
    st for st in
    block_exp.filter([{'sua':True}, {'unit_id':1}]) if len(st) > 2]
```

Create binned representation of experimental spike trains.

```python
sts_exp_bin = elephant.conversion.Binned(
    sts_exp, binsize=bin_size,
    t_start=rec_start, t_stop=rec_start + duration)
```

```python
print("Number of experimental spiketrains: " + str(len(sts_exp)))
```

## Load model data

Load all model simulation data of layer 5.

```python
filename = 'data/model.h5'
session = neo.NeoHdf5IO(filename=filename)
block_mdl = session.read_block()
```

Select spike trains (first exc. unit per recording channel).

```python
sts_mdl = []
for st in sts_exp:
   sts_mdl.append(block_mdl.filter(
       [{'channel_id': st.annotations['ca_id']}])[0])
```

Create binned representation of model spike trains.

```python
sts_mdl_bin = elephant.conversion.Binned(
    sts_mdl, binsize=bin_size,
    t_start=rec_start, t_stop=rec_start + duration)
```

Create binned representation of spike trains.

```python
print("Number of simulation spiketrains: " + str(len(sts_mdl)))
```

## Plot population activity rasters

Cut spike trains to 10 s of activity of plotting purposes.

```python
sts_exp_cut = [st.time_slice(10. * s, 20. * s) for st in sts_exp]
sts_mdl_cut = [st.time_slice(10. * s, 20. * s) for st in sts_mdl]
```

Plot 10 seconds of population activity as activity raster.

```python
plt.figure(figsize=(16,8))
plt.suptitle('Population Spike Activity', fontsize=20)

plt.subplot(2,1,1)
plt.title('Experiment', fontsize=18)
plots.spike_raster(sts_exp_cut)

plt.subplot(2,1,2)
plt.title('Model', fontsize=18)
plots.spike_raster(sts_mdl_cut)

plt.subplots_adjust(hspace=0.5, top=0.85)

plt.savefig('pop_raster.png', transparent=True)
```

## Population histograms

Plot corresponding population histograms for 10 seconds.

```python
phist_exp = elephant.statistics.peth(sts_exp_cut, 100. * ms, output='rate')
phist_mdl = elephant.statistics.peth(sts_mdl_cut, 100. * ms, output='rate')
```

```python
plt.figure(figsize=(16,6))
plt.suptitle('Population Spike Rate', fontsize=20)

plt.subplot(2,1,1)
plt.title('Experiment', fontsize=18)
plots.population_rate(phist_exp, col=exp_col)

plt.subplot(2,1,2)
plt.title('Model', fontsize=18)
plots.population_rate(phist_mdl, col=mdl_col)

plt.subplots_adjust(hspace=0.5, top=0.85)

plt.savefig('pop_peth.png', transparent=True)
```

## Cross-correlograms

Load cross-correlations calculated by a UP task, and compare the area under the central peak to the 5% significance level determined by spike jitter surrogates. The following plots show 5 representative examples of the CCH from the two datasets.

```python
plt.figure(num=None, figsize=(15, 12), dpi=120, facecolor='w', edgecolor='k')
plt.suptitle('Example cross-correlograms', fontsize=20)

filename = [
    '../results/hbp_review_task_L4/viz_output_exp.pkl',
    '../results/hbp_review_task_L4/viz_output_mdl.pkl']
titles = ['Experiment', 'Model']

num_cc_plots = 5

# Loop over experiment and model
for i in range(2):
    # load results from UP task
    f = open(filename[i], 'r')
    cc = pickle.load(f)
    f.close()

    # example: build correlation matrix
    num_neurons = cc['meta']['num_neurons']
    num_edges = cc['meta']['num_edges']
    
    for j in range(num_cc_plots):
        ax = plt.subplot(num_cc_plots, 2, i + 1 + j * 2)
        
        plots.cc_examples(cc, j, num_cc_plots)
        if j == 0:
            plt.title(titles[i], fontsize=18)
                
plt.subplots_adjust(hspace=1.0, top=0.9)

plt.savefig('cch_example.png', transparent=True)
```

```python
plt.figure(num=None, figsize=(15, 7), dpi=120, facecolor='w', edgecolor='k')
plt.suptitle('Significance of central peak in cross-correlation', fontsize=20)

# Loop over experiment and model
for i in range(2):
    # load results from UP task
    f = open(filename[i], 'r')
    cc = pickle.load(f)
    f.close()

    num_neurons = cc['meta']['num_neurons']
    num_edges = cc['meta']['num_edges']

    # Construct significance matrix (1-p) from results
    C = np.zeros((num_neurons, num_neurons))
    x = cc['edges']['id_i'].astype(int)
    y = cc['edges']['id_j'].astype(int)
    p = cc['func_conn']['cch_peak']['pvalue'] / 1000.
    for edge_i in range(num_edges):
        C[x[edge_i], y[edge_i]] = C[y[edge_i], x[edge_i]] = 1.-p[edge_i]

    ax = plt.subplot(1, 2, i + 1,aspect=1)
    plt.title(titles[i], fontsize=18)
    plots.cc_sig_matrix(C, num_neurons)

plt.subplots_adjust(hspace=1.0, top=0.88)

plt.savefig('cch_matrix.png', transparent=True)
```
