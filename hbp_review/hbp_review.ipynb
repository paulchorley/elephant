## Cross-correlograms

Load cross-correlations calculated by a UP task, and compare the area under the central peak to the 5% significance level determined by spike jitter surrogates.

```python
plt.figure(num=None, figsize=(15, 12), dpi=120, facecolor='w', edgecolor='k')
plt.suptitle('Example cross-correlograms', fontsize=20)

filename = [
    '../results/hbp_review_task/viz_output_exp.pkl',
    '../results/hbp_review_task/viz_output_mdl.pkl']
titles = ['Experiment', 'Model']
rdm=341
rdo=3

num_cc_plots = 5
for i in range(2):
    # load results from UP task
    f = open(filename[i], 'r')
    cc = pickle.load(f)
    f.close()

    # example: build correlation matrix
    num_neurons = cc['meta']['num_neurons']
    num_edges = cc['meta']['num_edges']
    
    k=np.zeros((10,10))

    for j in range(num_cc_plots):
        y = cc['edge_time_series']['cch'][j * rdm + rdo, :]
        x = cc['edge_time_series']['times_ms'][j * rdm + rdo, :]
        s1 = cc['edge_time_series']['sig_upper_975'][j * rdm + rdo, :]
        s2 = cc['edge_time_series']['sig_lower_25'][j * rdm + rdo, :]

        n1 = cc['edges']['id_i'][j * rdm + rdo]
        n2 = cc['edges']['id_j'][j * rdm + rdo]
        x1 = cc['neuron_topo']['x'][n1]
        y1 = cc['neuron_topo']['y'][n1]
        x2 = cc['neuron_topo']['x'][n2]
        y2 = cc['neuron_topo']['y'][n2]
        
        k[x1][y1]=1
        k[x2][y2]=1


        ax = plt.subplot(num_cc_plots, 2, i + 1 + j * 2)
        ind = np.argmin(np.abs(x))
        ax.fill_between(x[ind-5:ind+5], s1[ind-5:ind+5], color=[0.6, 0.6, 0.6])
        plt.plot(x, y, 'k-')
        if j == 0:
            plt.title(titles[i], fontsize=18)
        plt.ylabel(
            "El: %i - %i \n (%i,%i) - (%i,%i)" %
            (n1, n2, x1, y1, x2, y2),
            fontsize=16)
        if j == num_cc_plots - 1:
            plt.xlabel("time (ms)", fontsize=16)
        else:
            ax.set_xticklabels([])
        ax.tick_params(axis='both', which='major', labelsize=14)
        plt.axis('tight')
        plt.xlim((-100,100))
    plt.figure(num=None, figsize=(15, 12), dpi=120, facecolor='w', edgecolor='k')    
    
    plt.pcolor(np.arange(10), np.arange(10), k)
    plt.figure(num=None, figsize=(15, 12), dpi=120, facecolor='w', edgecolor='k')    
        
        
plt.savefig('cch_example.png', transparent=True)
```

```python
plt.figure(num=None, figsize=(15, 7), dpi=120, facecolor='w', edgecolor='k')
plt.suptitle('Significance of central peak in cross-correlation', fontsize=20)

for i in range(2):
    # load results from UP task
    f = open(filename[i], 'r')
    cc = pickle.load(f)
    f.close()

    num_neurons = cc['meta']['num_neurons']
    num_edges = cc['meta']['num_edges']

    C = np.zeros((num_neurons, num_neurons))
    x = cc['edges']['id_i'].astype(int)
    y = cc['edges']['id_j'].astype(int)
    p = cc['func_conn']['cch_peak']['pvalue'] / 1000.
    for edge_i in range(num_edges):
        C[x[edge_i], y[edge_i]] = C[y[edge_i], x[edge_i]] = p[edge_i]

    ax = plt.subplot(1, 2, i + 1,aspect=1)
    plt.pcolor(np.arange(num_neurons), np.arange(num_neurons), 1. - C)
    plt.title(titles[i], fontsize=18)
    plt.xlabel("Neuron ID i", fontsize=16)
    if i==0:
        plt.ylabel("Neuron ID j", fontsize=16)
    plt.axis('tight')
    ax.tick_params(axis='both', which='major', labelsize=14)
    plt.clim(0, 1)
    
    cb=plt.colorbar()
    cb.set_label('significance (1-p)', fontsize=14, labelpad=10, y=0.45)

plt.subplots_adjust(hspace=1.0, top=0.88)

plt.savefig('cch_matrix.png', transparent=True)
```
